# Linear Regression from Scratch with Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Documents/machine-learning-ex1/ex1/ex1data1.txt',names = ['X','y'])
```

```python
data.head()
```

```python
plt.plot(data['X'],data['y'],'bo')
```

```python
def cost_function(X,y,theta):
    cost = (1/2)*np.mean(((X@theta.T) - y)**2)
    print("Cost with theta [{:.2f},{:.2f}] = {:.2f}".format(theta[0,0],theta[0,1],cost))
    
```

```python
def gradientDescent(X,y,theta,alpha,iters):
    history = np.zeros((iters,1))
    for i in range(iters):
        delta = np.mean(((X@theta.T)-y)*X,axis = 0)
        theta = theta - alpha*delta
        history[i] = cost_function(X,y,theta)
    return theta
```

```python
def linear_regression(X, y, alpha,iters):
    
    theta = np.zeros((1,2))
    ones = np.ones((97,1))
    x = X.values.reshape(-1,1)
    X = np.append(ones,x, axis = 1)
    y = y.values.reshape(-1,1)
    
    cost_function(X,y,theta)
    theta = gradientDescent(X,y,theta,alpha,iters)
    return theta
```

```python
coef = linear_regression(data['X'],data['y'],0.02,1500)
```

```python
coef
```

```python
coef.shape
```

```python
data['intercept'] = 1
```

```python
X = data[['intercept','X']].values

yhat = X@coef.T
```

```python
plt.plot(X[:,1],yhat)
plt.plot(X[:,1],data['y'],'ro')
mse = np.mean((data['y'].values.reshape(-1,1)-yhat)**2)
print('mean average squared error = {:.2f}'.format(mse))
```

```python
def normal_eq(X,y):
    y = y.values.reshape(-1,1)
    
    theta = np.linalg.inv((X.T@X))@(X.T@y)

    return theta
```

```python
theta1 = normal_eq(X,data['y'])
theta1
```

```python
#Since it's already a 2x1 vector, it won't be transposed to perform the linear regression
yhat1 = X@theta1
```

```python
plt.plot(X[:,1],yhat1)
plt.plot(X[:,1],data['y'],'ro')
mse1 = np.mean((data['y'].values.reshape(-1,1)-yhat1)**2)
print('mean squared error = {:.2f}'.format(mse1))
```
